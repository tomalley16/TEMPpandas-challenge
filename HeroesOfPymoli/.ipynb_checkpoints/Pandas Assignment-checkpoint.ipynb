```python
# average purchase prise for each gender                -  GOOD
# get price for each gender and then the mean()
avg_male_purchase_price = male_purchase_data_df["Price"].mean()
avg_female_purchase_price = female_purchase_data_df["Price"].mean()
avg_other_purchase_price = other_purchase_data_df["Price"].mean()
```

```python
#total purchase value by gender        -GOOD
# sum the total for each 
male_purchase_total = sum(male_purchase_data_df["Price"])
female_purchase_total = sum(female_purchase_data_df["Price"])
other_purchase_total = sum(other_purchase_data_df["Price"])
```

```python
# avg purcahse total per person    # --- GOOD
# need sum of purchase total / total of each gender  # get the len()
avg_male_purchase_total = male_purchase_total/len(total_male_players)
avg_female_purchase_total = female_purchase_total/len(total_female_players)
avg_other_purchase_total = other_purchase_total/len(total_other_players)
```

```python
# Set up the dislplay     # --- GOOD
# set up index =[], money with "xxxx":"${:,.2f}"

index = ["Male", "Female", "Other"]
purchase_analysis = pd.DataFrame({"Purchase Count": [male_purchase_count,female_purchase_count, other_purchase_count],
                                 "Average Purchase Price": [avg_male_purchase_price, avg_female_purchase_price, avg_other_purchase_price],
                                 "Total Purchase Value": [male_purchase_total, female_purchase_total, other_purchase_total]},
                                index = index)
#purchase_analysis

# fix the cells
purchase_analysis.style.format({"Average Purchase Price": "${:,.2f}",
                                "Total Purchase Vale": "${:, .2f}"})
purchase_analysis
```
